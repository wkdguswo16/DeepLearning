```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
x = df.drop(['class'], axis=1).values
# x = x.reshape(x.shape[0], 3, 2)
y = df['class']
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.25, random_state=0)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
classes = df_origin['class'].value_counts().keys()
classes
```

```python
from typing import Callable
from sklearn.metrics import accuracy_score
from mlxtend.plotting import plot_confusion_matrix
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC

def show_matrix(predict: Callable, x, y, classes):
    y_pred = predict(x)
    score = accuracy_score(y, y_pred)
    mat = confusion_matrix(y, y_pred)
    plot_confusion_matrix(conf_mat=mat, class_names=classes,
                        show_normed=True, figsize=(7, 7))
    plt.title(f"{predict.__self__.__class__.__name__} score: {score*100:.2f}%")
clf_svm = SVC(random_state=0)
clf_svm.fit(x_train, y_train)

show_matrix(clf_svm.predict, x_test, y_test, classes)
```

```python
from sklearn.linear_model import LogisticRegression

clf_lr = LogisticRegression(random_state=0)
clf_lr.fit(x_train, y_train)

pred_lr = clf_lr.predict(x_test)

show_matrix(clf_lr.predict, x_test, y_test, classes)
```

```python
from sklearn.tree import DecisionTreeClassifier

clf_dt = DecisionTreeClassifier(random_state=0)
clf_dt.fit(x_train, y_train)

pred_dt = clf_dt.predict(x_test)

show_matrix(clf_dt.predict, x_test, y_test, classes)
```

```python
from sklearn.ensemble import RandomForestClassifier

print ("\n--- Random Forest ---")
rf_clf = RandomForestClassifier(random_state=0)
rf_clf.fit(x_train, y_train)

show_matrix(rf_clf.predict, x_test, y_test, classes)
```
